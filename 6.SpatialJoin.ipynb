```python
import pandas as pd
import geopandas as gpd
from keplergl import KeplerGl
```

```python
gdf = gpd.read_file('data\parking_2015.geojson')
gdf.head()
```

```python
nbs = gpd.read_file(r'data\neighbrhoods.geojson')
nbs.head()
```

```python
gdf.crs == nbs.crs
```

```python
nbs = nbs.to_crs(epsg=4326)
```

```python
gdf.crs == nbs.crs
```

```python
location_join = gpd.sjoin(gdf, nbs, op='within')
```

```python
location_join.head()
```

```python
location_join.sample(10)
```

```python
def count_numpoints(joined, polygons, merge_on):
    grouped = joined.groupby(merge_on).size()
    df = grouped.to_frame().reset_index()
    df.columns = [merge_on, 'Count']
    return polygons.merge(df, on=merge_on, how='outer')

numpoints_per_neigh = count_numpoints(location_join, nbs, 'NAME')
numpoints_per_neigh.head()
```

```python
m = KeplerGl(height=900)
m.add_data(data=numpoints_per_neigh, name='choropleth_map')
m
```
